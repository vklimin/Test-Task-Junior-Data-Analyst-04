```python
import pandas as pd
```

```python
# Загрузка данных во фреймы
# Финансовая информация
df_financial_data = pd.read_csv("../data/financial_data.csv")
# Информация о пролонгациях
df_prolongations = pd.read_csv("../data/prolongations.csv")
```

```python
# Визуальное ознакомление с данными о пролонгациях
df_prolongations.head()
```

```python
# Унификация данных о датах
MNAME_TO_NUM = {
    "январь": "01",
    "февраль": "02",
    "март": "03",
    "апрель": "04",
    "май": "05",
    "июнь": "06",
    "июль": "07",
    "август": "08",
    "сентябрь": "09",
    "октябрь": "10",
    "ноябрь": "11",
    "декабрь": "12",
}


def date_parse(name):
    parts = name.strip().lower().split()
    if (len(parts) == 2) and (parts[0] in MNAME_TO_NUM) and parts[1].isdigit():
        return f"{parts[1]}-{MNAME_TO_NUM[parts[0]]}"
    return name


df_financial_data.columns = df_financial_data.columns.map(date_parse)
df_prolongations["month"] = df_prolongations["month"].map(date_parse)
```

```python
# Проверка отсутствия пропущенных значений в id
if df_financial_data["id"].isna().any():
    print(f'Столбец "id" в financial_data содержит значения NaN')

if df_prolongations["id"].isna().any():
    print(f'Столбец "id" в prolongations содержит значения NaN')
```

```python
if df_financial_data["Причина дубля"].isna().any():
    df_financial_data["Причина дубля"] = df_financial_data["Причина дубля"].fillna("Полная оплата")
    print(f'Столбец "Причина дубля" в financial_data содержал значения NaN')
    print('Скорректировано на "Полная оплата"')

# Список причин дублей для ознакомления
df_financial_data["Причина дубля"].unique().tolist()
# Считаем, что "Полная оплата" означает проект без дублей оплат
# В ТЗ не определено наше отношение в причинам дублей, считаем их равноценными.
```

```python
# Проверка отсутствия пропущенных значений в month
if df_prolongations["month"].isna().any():
    print(f'Столбец "month" в prolongations содержит значения NaN')
```

```python
# Проверка наличия проектов без первичных данных о менеджере и дате завершения
df_id_diff = set(df_financial_data["id"]) - set(df_prolongations["id"])
if df_id_diff != set([]):
    # У нас есть проект без явно назначенного менеджера в первичном источнике данных
    print(
        f'Столбец "id" в financial_data содержит значения {df_id_diff} без пары в первичном источнике (prolongations)'
    )
    # Значит имя менеджера нужно будет брать из вторичных данных
    # Более того, из этого вытекает, что дата завершения проекта не определена
```

```python
base_columns = ["id", "Причина дубля", "Account"]
date_columns = sorted(set(df_financial_data.columns) - set(base_columns))
```

```python
# Подготовка данных в столбцах с финансовой информацией
df_financial_data_cleaned = df_financial_data.copy()

for column in date_columns:
    df_financial_data_cleaned[column] = pd.to_numeric(
        df_financial_data_cleaned[column]
        .str.replace(",", ".", regex=False)
        .str.replace("\xa0", "", regex=False)
        .str.replace("в ноль", "0.0", regex=False)
        .str.replace("стоп", "-1", regex=False)
        .str.replace("end", "-1", regex=False)
    )

df_financial_data_cleaned
```

```python
# Устранение дублирующейся информации по менеджерам и ведомым проектам
df_prolongations_cleaned = df_prolongations.drop_duplicates().reset_index(drop=True)
print(f'Устранено дублирующихся записей: {len(df_prolongations) - len(df_prolongations_cleaned)} шт.')
```

```python
# Агрегация данных по каждому "проекту-менеджеру"
df_financial_data_grouped = (
    df_financial_data_cleaned.groupby(["id", "Account"])
    .agg(lambda x: x.dropna().tolist())
    .reset_index()
)

df_financial_data_grouped
```

```python
# Объединение данных таблиц с проектами и менеджерами для дальнейшего анализа
df_data = df_financial_data_grouped.merge(
    df_prolongations_cleaned, on=["id"], how="inner"
)

# Для случая NaN в столбце "AM" по ТЗ информация берётся из вторичного источника (столбец "Account")
mask = df_data["AM"].isna()
if mask.any():
    df_data.loc[mask, "AM"] = df_data.loc[mask, "Account"]
    print('Столбец "AM" содержал значения NaN')
    print('Скорректировано на значение из вторичного истоника - столбца "Account"')

df_data = df_data.drop(columns=["Account", "Причина дубля"])

df_data
```

```python
# Исключение из расчётов данных об остановленных проектах
skipped = []
skipped_data = []
for idx, row in df_data.iterrows():
    for c in date_columns:
        if c <= row["month"] and (-1 in row[c]):
            skipped.append(idx)
            skipped_data.append((row["id"], row["month"]))

print(f"Обнаружено {len(skipped)} остановок проектов: {skipped_data}")
df_data_cleaned = df_data[~df_data.index.isin(skipped)].copy()
print("Записи об остановленных проектах исключены из расчётов")

df_data_cleaned
```

```python
# Сбор статистики по продлению договоров
PROLONGATION_MONTH_COL = "Дата пролонгации"
K1_DENOMINATOR = "Знаменатель K1"  # Сумма отгрузки в последний месяц проекта
K1_NUMERATOR = "Числитель K1"  # Сумма отгрузки в первый месяц после завершения (своевременная пролонгация)
K2_DENOMINATOR = "Знаменатель K2"  # Сумма отгрузки в последний месяц проекта, но непролонгированного своевременно в первый месяц
K2_NUMERATOR = "Числитель K2"  # Сумма отгрузки во второй месяц после завершения при нулевой в первый месяц (несвоевременная пролонгация)


def get_metrics(row):
    prolongation_month = "Непролонгирован"
    k1_denominator = 0.0
    k1_numerator = 0.0
    k2_denominator = 0.0
    k2_numerator = 0.0

    # Последний месяц проекта
    project_end = row["month"]
    
    # Список месяцев до окончания проекта (включительно)
    past_columns = [c for c in date_columns if c <= project_end]
    # Список месяцев после окончания проекта
    future_columns = [c for c in date_columns if c > project_end]

    # Сумма отгрузки в последний НЕНУЛЕВОЙ месяц проекта
    if len(past_columns) > 0:
        k1_denominator = sum(row[past_columns[-1]])
        # Если отгрузка в последний месяц "в ноль",
        if (k1_denominator == 0) and len(past_columns) > 1:  # то берем предыдущий месяц
            k1_denominator = sum(row[past_columns[-2]])

    # Если проект завершился и сумма завершения не "в ноль"
    if k1_denominator > 0:
        # Считаем выручку в первом месяце после завершения
        k1_numerator = (
            sum(value for value in row[future_columns[0]] if value > 0)
            if len(future_columns) > 0
            else 0
        )
        
        # Если выручка была (проект своевременно продлён)
        if k1_numerator > 0:
            # то фиксируем месяц продления
            prolongation_month = future_columns[0]
        else:  # Иначе
            # Фиксируем сумму завершившегося, но не пролонгированного в первый месяц проекта
            k2_denominator = k1_denominator
            # И пробуем посмотреть на второй месяц после завершения проекта
            k2_numerator = (
                sum(value for value in row[future_columns[1]] if value > 0)
                if len(future_columns) > 1
                else 0
            )
            # Если выручка пошла (пролонгирован, хть и с опозданием)
            if k2_numerator > 0:
                # тогда месяцем продления будем считать второй месяц
                prolongation_month = future_columns[1]

    return (
        prolongation_month,
        k1_denominator,
        k1_numerator,
        k2_denominator,
        k2_numerator,
    )


# Сбор статистики по пролонгациям по каждому проекту в отдельности
df_data_cleaned[
    [
        PROLONGATION_MONTH_COL,
        K1_DENOMINATOR,
        K1_NUMERATOR,
        K2_DENOMINATOR,
        K2_NUMERATOR,
    ]
] = df_data_cleaned.apply(get_metrics, axis=1, result_type="expand")

df_data_cleaned
```

```python
# Выгрузка данных в промежуточный документ
# для дальнейшей компановки и визуализации в Excel
df_data_cleaned[["AM", PROLONGATION_MONTH_COL, K1_DENOMINATOR, K1_NUMERATOR, K2_DENOMINATOR, K2_NUMERATOR]].to_excel("output.xlsx", index=False)
```
